```python
%sql
--Criando um schema só para a camada bronze
Create DATABASE Bronze
```

```python
#Leitura da tabela Order
df_order = spark.read.json("dbfs:/Volumes/workspace/ifood/base_case_ifood/order.json.gz")

#Carga de dados na camada Bronze
df_order.write.format("delta").mode("overwrite").saveAsTable("Bronze.orders_raw")
```

```python
# dados de consumidores
df_users_raw = spark.read.option("header", True).csv("dbfs:/Volumes/workspace/ifood/base_case_ifood/consumer.csv.gz")

#persistindo os dados em tabelas
df_users_raw.write.format("delta").mode("overwrite").saveAsTable("Bronze.users_raw")
```

```python
# dados de Restaurantes
df_restaurant_raw = spark.read.option("header", True).csv("dbfs:/Volumes/workspace/ifood/base_case_ifood/restaurant.csv.gz")

#persistindo os dados em tabelas
df_restaurant_raw.write.format("delta").mode("overwrite").saveAsTable("Bronze.restaurant_raw")
```

```python
# dados de Teste AB
df_ab_raw = spark.read.option("header", True).csv("dbfs:/Volumes/workspace/ifood/base_case_ifood/ab_test_ref.csv")

#persistindo os dados em tabelas
df_ab_raw.write.format("delta").mode("overwrite").saveAsTable("Bronze.ab_test_ref")
```
